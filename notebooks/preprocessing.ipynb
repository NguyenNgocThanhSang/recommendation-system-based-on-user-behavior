## Xử lý clickstream data

```python
clickstream = pd.read_csv('../data/raw/click_stream.csv')
clickstream.head(5)
```

```python
# lọc bỏ những dòng có customer_id bị khuyết
clickstream.dropna(subset=['customer_id'], inplace=True)
```

```python
clickstream['session_id'].isnull().sum()
```

```python
clickstream['event_name'].isnull().sum()
```

```python
clickstream['event_name'].unique()
```

```python
# loại bỏ những hành động không thể hiện tương tác của người dùng đối với sản phẩm
actions = ['ADD_TO_CART', 'BOOKING']

clickstream = clickstream[clickstream['event_name'].isin(actions)]
clickstream
```

```python
clickstream['event_time'].isnull().sum()
```

```python
# chuyển dữ liệu event_time sang kiểu datetime
clickstream['event_time'] = pd.to_datetime(clickstream['event_time'])
```

```python
# sắp xếp lại thời gian sự kiện
clickstream = clickstream.sort_values(by='event_time')
clickstream
```

```python
# loại bỏ cột event_id
clickstream.drop(columns='event_id', inplace=True)
clickstream
```

```python
# loại bỏ cột traffic_source
clickstream.drop(columns='traffic_source', inplace=True)
clickstream
```

```python
# loại bỏ cột event_matadata
clickstream.drop(columns='event_metadata', inplace=True)
clickstream
```

```python
# xóa cột payment_status
clickstream.drop(columns='payment_status', inplace=True)
clickstream
```

```python
# xóa cột search_keywords
clickstream.drop(columns='search_keywords', inplace=True)
clickstream
```

```python
# xóa booking_id
clickstream.drop(columns='booking_id', inplace=True)
clickstream
```

```python
# bỏ 2 cột cuối cùng là item_price và price_usd
clickstream.drop(columns=['item_price', 'price_usd'], inplace=True)
clickstream
```

```python
clickstream['product_id'].isna().sum()
```

```python
# reset index của bộ dữ liệu
clickstream.reset_index(inplace=True)
clickstream
```

```python
clickstream.info()
```

```python
clickstream.columns
```

```python
# Lưu lại vào file event.csv
clickstream.to_csv('../data/processed/event.csv', index=False)
```

## Xử lý product data

```python
product = pd.read_csv('../data/raw/product.csv')
product
```

```python
product.isna().sum()
```

```python
product['basecolour'].unique()
```

```python
product[(product['basecolour'].isna()) & (product['articletype']=='Jeans')]
```

```python
import re

# Hàm tìm màu sắc từ productdisplayname
def extract_colour(name):
    # Danh sách các màu cần tìm
    colours = ["Blue", "Black", "White", "Red", "Green", "Yellow", "Grey", "Brown", "Pink", "Purple", "Orange"]
    # Tìm màu trong chuỗi
    for colour in colours:
        if re.search(rf'\b{colour}\b', name, re.IGNORECASE):  # So khớp không phân biệt hoa thường
            return colour
    return None

# Áp dụng cập nhật cột basecolour
product['basecolour'] = product.apply(
    lambda row: extract_colour(row['productdisplayname']) if pd.isna(row['basecolour']) else row['basecolour'], axis=1
)

product[(product['basecolour'].isna()) & (product['articletype']=='Jeans')]
```

```python
product['basecolour'].fillna("None", inplace=True)
```

```python
product['season'].fillna("None", inplace=True)
```

```python
product[product['year'].isna()]
```

```python
product['year'] = product['year'].fillna(2012)
```

```python
product['usage'] = product['usage'].fillna('Casual')
```

```python
product[product['usage'].isna()]
```

```python
product.isna().sum()
```

```python
product.info()
```

```python
# lưu lại vào file product.csv
product.to_csv('../data/processed/product.csv', index=False)
```
